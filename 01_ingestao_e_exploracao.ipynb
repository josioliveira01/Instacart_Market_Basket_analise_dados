```python
import pandas as pd
```

**Carregar para a memória as tabelas que foram extraidas da base do Kaggle**

```python

orders = pd.read_csv(
    r"C:\Users\olive\OneDrive\Documentos\Dio\Portfólio\analise-compras\data\raw\orders.csv"
)

order_prior = pd.read_csv(
    r"C:\Users\olive\OneDrive\Documentos\Dio\Portfólio\analise-compras\data\raw\order_products__prior.csv"
)

products = pd.read_csv(
    r"C:\Users\olive\OneDrive\Documentos\Dio\Portfólio\analise-compras\data\raw\products.csv"
)

aisles = pd.read_csv(
    r"C:\Users\olive\OneDrive\Documentos\Dio\Portfólio\analise-compras\data\raw\aisles.csv"
)

departments = pd.read_csv(
    r"C:\Users\olive\OneDrive\Documentos\Dio\Portfólio\analise-compras\data\raw\departments.csv"
)
order_train = pd.read_csv(
    r"C:\Users\olive\OneDrive\Documentos\Dio\Portfólio\analise-compras\data\raw\order_products__train.csv"
)
```

**Unindo as duas planilhas prior e train para contabilizar todos os clientes que foram separados em treino e teste**

```python
# 2. Unindo as duas tabelas (empilhamento vertical)
# O ignore_index=True garante que o novo DataFrame tenha um índice único e contínuo
df_produtos_total = pd.concat([order_prior,order_train], ignore_index=True)

# 3. Verificação
print(f"Linhas em Prior: {len(order_prior)}")
print(f"Linhas em Train: {len(order_train)}")
print(f"Total após a união: {len(df_produtos_total)}")


# 3. Salvando o arquivo final na pasta 'processed'
df_produtos_total.to_csv(
    r"C:\Users\olive\OneDrive\Documentos\Dio\Portfólio\analise-compras\data\raw\order_products_total.csv",
    index=False
)

print(f"Sucesso! A base unificada com {len(df_produtos_total)} linhas foi salva.")
```
