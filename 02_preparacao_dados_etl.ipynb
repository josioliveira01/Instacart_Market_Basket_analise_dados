Objetivo: Realizar o processo de ETL (Extração, Transformação e Carga), integrando as bases do Instacart e preparando uma base analítica consistente para análise de comportamento de compra dos clientes.

```python
import pandas as pd
```

Carregar os dados da camada raw sem modificá-los.

```python
orders = pd.read_csv(
    r"C:\Users\olive\OneDrive\Documentos\Dio\Portfólio\analise-compras\data\raw\orders.csv"
)

order_prior = pd.read_csv(
    r"C:\Users\olive\OneDrive\Documentos\Dio\Portfólio\analise-compras\data\raw\order_products__prior.csv"
)

products = pd.read_csv(
    r"C:\Users\olive\OneDrive\Documentos\Dio\Portfólio\analise-compras\data\raw\products.csv"
)

aisles = pd.read_csv(
    r"C:\Users\olive\OneDrive\Documentos\Dio\Portfólio\analise-compras\data\raw\aisles.csv"
)

departments = pd.read_csv(
    r"C:\Users\olive\OneDrive\Documentos\Dio\Portfólio\analise-compras\data\raw\departments.csv"
)

order_train = pd.read_csv(
    r"C:\Users\olive\OneDrive\Documentos\Dio\Portfólio\analise-compras\data\raw\order_products__train.csv"
)
```

```python
print(orders.info())
```

```python
orders.isnull().sum()
```

*Integração das tabelas*
Pedidos + produtos comprados

```python
df = order_prior.merge(
    orders,
    on="order_id",
    how="left"
)
```

Adicionar informações de produtos

```python
df = df.merge(
    products,
    on="product_id",
    how="left"
)
```

Adicionar categorias (aisles e departments)

```python
df = df.merge(
    aisles,
    on="aisle_id",
    how="left"
)

df = df.merge(
    departments,
    on="department_id",
    how="left"
)
```

Seleção de colunas relevantes

```python
df = df[
    [
        "user_id",
        "order_id",
        "order_number",
        "order_dow",
        "order_hour_of_day",
        "days_since_prior_order",
        "product_id",
        "product_name",
        "aisle",
        "department",
        "add_to_cart_order",
        "reordered"
    ]
]
```
